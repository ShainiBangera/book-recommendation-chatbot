```python
#intents
intents = {
    "greeting": {
        "patterns": ["hi", "hello", "hey"],
        "responses": ["Hello!", "Hi there!", "Hey!"]
    },
    "book_query": {
        "patterns": ["book", "recommend", "suggest"],
        "responses": ["Please tell me the programming language."]
    },
    "return_book": {
        "patterns": ["return", "give back", "return book"],
        "responses": ["Which book do you want to return?"]
    },
    "fine_query": {
        "patterns": ["fine", "late"],
        "responses": ["Please tell me number of late days."]
    },
    "goodbye": {
        "patterns": ["bye", "exit"],
        "responses": ["Goodbye!", "See you later!"]
    }
}

#Book categories (programming Languages)
categories ={
    "ai" : ["Artificial Intelligence For Dummies (2nd Edition)", "AI and Machine Learning for Coders", "Programming PyTorch for Deep Learning", "Introduction to Machine Learning with Python", "AI Engineering"],
    "python" : ["Python Crash Course", "Fluent Python", "Python in a Nutshell", "Python Cookbook", "Python for Data Analysis"],
    "java" : ["Java: A Beginner's Guide", "Java: The Complete Reference","Effective Java","Core Java Volume 1-Fundamentals", "Head First Java"],
    "c" : ["C Programming Absolute Beginner's Guide", "The C Programming Language", "C Programming: A Modern Approach", "Head First C", "C: The Complete Reference"],
    "c++" : ["C++ Programming: The Big Three", "A Complete Guide to Programming in C++", "C++ Primer", "C++ How to Program", "Object Oriented Programming in C++", "Effective C++"],
    "html and css" : ["Head First HTML and CSS", " HTML and CSS Quickstart Guide", "Learning Web Design", "Responsive Web Design with HTML5 and CSS", "HTML and CSS: The Comprehensive Guide"],
    "javascript" : ["A Smarter Way to Learn JavaScript", "JavaScript from Beginner to Professional", "JavaScript: The Good Parts", "JavaScript: The Definitive Guide", "Eloquent JavaScript"],
    "sql" : ["Learning SQL", "Getting Started with SQL", "SQL All-in-One For Dummies", "SQL Cookbook", "SQL for Smarties"]
}
```

```python
\import random

#intent detection function
def get_intent(user_input):
  user_input = user_input.lower() #converts it into lowercase

  if "return" in user_input or "give back" in user_input:
        return "return_book"

  for intent, data in intents.items():
    for pattern in data["patterns"]:
      #checks if any keyword is present in user input
      if pattern in user_input:
        return intent

  return "unknown"

#extract category function
def extract_category(user_input):
  user_input = user_input.lower()

  #check if any category name is present in input
  for category in categories:
    if category in user_input:
      return category

  return None  #if no category found

def calculate_fine(days):
    return days * 10

def chatbot():
  issued_books = []  #list to store issued books

  while True:
    try:
      user_input = input("User: ").lower()
      intent = get_intent(user_input) #detect intent

      #greeting
      if intent == "greeting":
        print("Bot: ", random.choice(intents[intent]["responses"]))

      #exit
      elif intent == "goodbye":
        print("Bot: ", random.choice(intents[intent]["responses"]))
        break  #stops chat


      #return book
      elif intent == "return_book":
        if not issued_books:
          print("Bot: No books issued to return.")
        else:
          print("Bot: Here are the books you have issued:")
          for b in issued_books:
            print("-", b)

          print("Bot: Enter book name to return:")
          book_name = input("User: ").lower()

          found = False

          #checking if book is in issued list
          for b in issued_books:
            if book_name == b.lower():
              issued_books.remove(b)
              print(f"Bot: '{b}' returned successfully.")

              print("Bot: Enter late days (0 if none):")
              days = int(input("User: "))

              fine = calculate_fine(days)

              if fine > 0:
                print(f"Bot: Fine is Rs.{fine}")
              else:
                print("Bot: No fine. Thank you!")

              found = True
              break

          else:
            print("Bot: This book was not issued by you.")

          print(f"Bot: Remaining books: {len(issued_books)}")

      elif intent == "book_query":
        #shows all categories
        print("Bot: Available categories:")
        for category in categories:
          print(f"- {category}")

        #asks user to choose category
        print("Bot: Please choose a category:")
        category_choice = input("User: ").lower()

        category = extract_category(category_choice)

        if category:
          print(f"Bot: Here are some books in {category}:")

          #display books in that category
          for b in categories[category]:
            print("-", b)

          #ask user to select books
          print("Bot: Please select books (comma separated):")
          selected = input("User: ").lower()
          selected_books = [book.strip() for book in selected.split(",")] #splits input into list

          valid_books = []
          invalid_books = []

          #check each selected book
          for book in selected_books:
            matched = False

            for actual_book in categories[category]:
              if book.strip().lower() == actual_book.strip().lower():
                valid_books.append(actual_book)  # keeps original format
                matched = True
                break

            if not matched:
              invalid_books.append(book)

          #if-else conditions
          if len(valid_books) > 3:
            print("Bot: You can issue maximum 3 books only!")
          elif len(valid_books) == 0:
            print("Bot: No books selected.")
          else:
            issued_books.extend(valid_books)
            print("Bot: Books issued successfully!")
            for b in valid_books:
              print("-", b)

          if invalid_books:
            print("Bot: These books are not available:")
            for b in invalid_books:
              print("-", b)

          #total books issued
          print(f"Bot: Total books issued: {len(issued_books)}")

        else:
         print("Bot: Invalid category.")


      #fine query
      elif intent == "fine_query":
        print("Bot: Enter number of late days:")
        days = int(input("User: "))

        fine = calculate_fine(days)
        print(f"Bot: Your fine is Rs.{fine}")

      else:
        print("Bot: I didn't understand.")

    except ValueError:
      print("Bot: Please enter valid numeric input!")  # if wrong number input
    except Exception as e:
      print("Bot: Something went wrong:", e)  # general error

chatbot()
```
